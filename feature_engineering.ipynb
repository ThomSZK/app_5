```python
# Função para carregar o arquivo JSON de habilidades
def load_skills_from_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        skills_data = json.load(f)
    
    # Extraindo habilidades e seus sinônimos
    known_skills = set()
    for skill in skills_data:
        known_skills.add(skill["nome"])  # Adicionando o nome da habilidade
        known_skills.update(skill["sinonimos"])  # Adicionando os sinônimos da habilidade
    
    return known_skills

def extract_skills_from_text(text, skills_data, threshold):
    # Preparar listas e mapas
    all_terms = []
    term_to_skill = {}

    text = normalize(text)
    
    for skill in skills_data:
        nome = normalize(skill["nome"])
        all_terms.append(nome)
        term_to_skill[nome] = skill["nome"]
        for sin in skill.get("sinonimos", []):
            sin_norm = normalize(sin)
            all_terms.append(sin_norm)
            term_to_skill[sin_norm] = skill["nome"]
    
    # Normalizar e codificar texto
    text_sentences = [normalize(s) for s in text.split('\n') if s.strip()]
    sentence_embeddings = model.encode(text_sentences, convert_to_tensor=True)
    skill_embeddings = model.encode(all_terms, convert_to_tensor=True)

    results = set()
    for sentence_embedding in sentence_embeddings:
        sim_scores = util.cos_sim(sentence_embedding, skill_embeddings)[0]
        for i, score in enumerate(sim_scores):
            if score > threshold:
                skill_name = term_to_skill[all_terms[i]]
                results.add(skill_name)

    return results
```

```python
from sentence_transformers import SentenceTransformer, util

# Carrega modelo leve e eficaz para comparação semântica
model = SentenceTransformer('all-MiniLM-L6-v2')

def calcular_similaridade_vaga_curriculo(texto_vaga, resumo_curriculo):
    if pd.isna(texto_vaga) or pd.isna(resumo_curriculo):
        return 0.0
    else:
        # Gera embeddings dos textos
        embedding_vaga = model.encode(texto_vaga, convert_to_tensor=True)
        embedding_curriculo = model.encode(resumo_curriculo, convert_to_tensor=True)

        # Calcula a similaridade de cosseno entre os vetores
        similaridade = util.cos_sim(embedding_vaga, embedding_curriculo)

        # Retorna valor como float
        return similaridade.item()
```

```python
def hierarquia_idioma(df):
    df['formacao_e_idiomas_nivel_ingles'] = df['formacao_e_idiomas_nivel_ingles'].fillna('')
    df['formacao_e_idiomas_nivel_espanhol'] = df['formacao_e_idiomas_nivel_espanhol'].fillna('')
    
    df['formacao_e_idiomas_nivel_ingles'] = df['formacao_e_idiomas_nivel_ingles'].replace({'': 0,'Nenhum': 0,'Básico': 1, 'Intermediário': 2, 'Avançado': 3, 'Fluente': 4,'Técnico': 5})
    df['formacao_e_idiomas_nivel_espanhol'] = df['formacao_e_idiomas_nivel_ingles'].replace({'': 0,'Nenhum': 0,'Básico': 1, 'Intermediário': 2, 'Avançado': 3, 'Fluente': 4,'Técnico': 5})
    
    df['perfil_vaga_nivel_ingles'] = df['perfil_vaga_nivel_ingles'].replace({'': 0,'Nenhum': 0,'Básico': 1, 'Intermediário': 2, 'Avançado': 3, 'Fluente': 4,'Técnico': 5})
    df['perfil_vaga_nivel_espanhol'] = df['perfil_vaga_nivel_espanhol'].replace({'': 0,'Nenhum': 0,'Básico': 1, 'Intermediário': 2, 'Avançado': 3, 'Fluente': 4,'Técnico': 5})

    return df

def avaliador_idioma_ingles(df):
    cond = df['formacao_e_idiomas_nivel_ingles'] >= df['perfil_vaga_nivel_ingles']
    df['avaliador_idioma_ingles'] = df['formacao_e_idiomas_nivel_ingles'] - df['perfil_vaga_nivel_ingles']
    df.loc[cond, 'avaliador_idioma_ingles'] = 1
    df.loc[~cond, 'avaliador_idioma_ingles'] = df.loc[~cond, 'avaliador_idioma_ingles'].abs()
    df.loc[~cond,'avaliador_idioma_ingles'] = df.loc[~cond,'avaliador_idioma_ingles'].replace({0: 1, 1: 0.5, 2: 0.25, 3: 0.125, 4: 0.0625, 5: 0.03125})
    return df
def avaliador_idioma_espanhol(df):
    cond = df['formacao_e_idiomas_nivel_espanhol'] >= df['perfil_vaga_nivel_espanhol']
    df['avaliador_idioma_espanhol'] = df['formacao_e_idiomas_nivel_espanhol'] - df['perfil_vaga_nivel_espanhol']
    df.loc[cond, 'avaliador_idioma_espanhol'] = 1
    df.loc[~cond, 'avaliador_idioma_espanhol'] = df.loc[~cond, 'avaliador_idioma_espanhol'].abs()
    df.loc[~cond,'avaliador_idioma_espanhol'] = df.loc[~cond,'avaliador_idioma_espanhol'].replace({0: 1, 1: 0.5, 2: 0.25, 3: 0.125, 4: 0.0625, 5: 0.03125})
    return df
    
```
