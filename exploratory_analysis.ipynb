```python
import pandas as pd
import numpy as np
import json
# Show all columns without truncation
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)
```

## Vagas analysis

```python
with open('data/vagas/vagas.json', 'r') as f:
    data = json.load(f)

# Flatten JSON
flattened_data = []
for key, value in data.items():
    row = {"id": key}
    for section, details in value.items():
        for sub_key, sub_value in details.items():
            row[f"{section}_{sub_key}"] = sub_value
    flattened_data.append(row)

vagas = pd.DataFrame(flattened_data)
```

```python
vagas.sample(5)
```

## applicants analysis

```python
with open('data/applicants/applicants.json', 'r') as f:
    data = json.load(f)

# Flatten JSON
flattened_data = []
for key, value in data.items():
    row = {"id": key}
    for section, details in value.items():
        if isinstance(details, dict):  # Check if 'details' is a dictionary
            for sub_key, sub_value in details.items():
                row[f"{section}_{sub_key}"] = sub_value
        else:  # If 'details' is not a dictionary, add it directly
            row[section] = details
    flattened_data.append(row)

applicants = pd.DataFrame(flattened_data)
```

```python
applicants.sample(5)
```

## Prospects analysis

```python
with open('data/prospects/prospects.json', 'r') as f:
    data = json.load(f)

rows = []
for job_id, job_data in data.items():
    titulo = job_data.get("titulo", "")
    modalidade = job_data.get("modalidade", "")
    prospects = job_data.get("prospects", [])
    
    for prospect in prospects:
        row = {
            "job_id": job_id,
            "titulo": titulo,
            "modalidade": modalidade,
            "nome": prospect.get("nome", ""),
            "codigo": prospect.get("codigo", ""),
            "situacao_candidado": prospect.get("situacao_candidado", ""),
            "data_candidatura": prospect.get("data_candidatura", ""),
            "ultima_atualizacao": prospect.get("ultima_atualizacao", ""),
            "comentario": prospect.get("comentario", ""),
            "recrutador": prospect.get("recrutador", "")
        }
        rows.append(row)

# Create a pandas DataFrame
prospects = pd.DataFrame(rows)
```
